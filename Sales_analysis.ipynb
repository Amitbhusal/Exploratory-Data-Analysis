```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("Diwali Sales Data.csv",encoding="unicode_escape")
#to avoid encoding error,use unicode escape 
df = pd.DataFrame(data)
print(df)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns = df.columns.str.strip()
df.drop(["Status","unnamed1"],axis =1,inplace =True,errors= "ignore")
print(df)
```

```python
df.info()
```

```python

df.isnull().sum()
```

```python
df.dropna(inplace = True)  #inplace true saves the data after changes 
```

```python
df.isnull().sum()
```

```python
df["Amount"] = df["Amount"].astype(int)
```

```python
df["Amount"].dtype
```

```python
df.describe()
```

```python
df[["Age","Orders","Amount"]].describe()
```

##  Exploratory Data Analysis
#### Gender

```python
df.columns
```

```python
a = sns.countplot(data = df, x ="Gender")
for bars in a.containers:
    a.bar_label(bars)
```

```python
q = df.groupby(["Gender"],as_index = False)["Amount"].sum().sort_values(by ="Amount" ,ascending =False)
print(q)
```

```python

sns.barplot(data = q , x = "Gender",y ="Amount",color ="green")
```


##### From above  graphs we can see that most of the buyers are Female and even the purchasing powers of the female are greater than the men


#### Age

```python
ax =sns.countplot(data =df ,x ="Age Group",hue ="Gender",palette ="spring")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
bx = df.groupby(["Age Group"],as_index =False)["Amount"].sum().sort_values(by ="Amount" ,ascending = False)
print(bx)
sns.barplot(data =bx ,x = "Age Group" ,y ="Amount",color ="brown")
```

#### From above graphs we can see that most of the buyers are females of age group 26-35

### State

```python
# total numbers of order from top 10 states 
cx = df.groupby(["State"],as_index = False)["Orders"].sum().sort_values(by ="Orders" ,ascending = False)
#print(cx)
dx = cx.head(10)
print(dx)
sns.set(rc = {"figure.figsize":(15,5)})
sns.barplot(data =dx ,x ="State" ,y="Orders",hue ="State")
```

```python
# total amounts/sales from top 10 states
ex = df.groupby(["State"],as_index = False)["Amount"].sum().sort_values(by ="Amount" ,ascending = False).head(10)
print(ex)
sns.set(rc = {"figure.figsize":(15,5)})
sns.barplot(data =ex ,x ="State" ,y="Amount",hue ="State")
```

#### From above graphs we can see most of the orders and most amount spent are from Uttar Pradesh, Maharastra,Karnataka.But we get some changes in top 10 orders and amount spent as Kerela from orders goes out and Bihar gets in top 10.

### Marital Status

```python
df.columns
```

```python
fx = sns.countplot(data = df ,x = "Marital_Status")

for bars in fx.containers:
    fx.bar_label(bars)
```

```python
gx = df.groupby(["Marital_Status","Gender"],as_index = False)["Amount"].sum()
print(gx)
sns.barplot(data= gx ,x ="Marital_Status" ,y ="Amount",hue ="Gender")
```

#### From above graph we can see that most of the buyers are married(woman) and they have more purchasing power.

### Occupation 

```python
sns.set(rc={"figure.figsize":(20,5)})
hx =sns.countplot(data =df ,x ="Occupation",hue ="Occupation")
for bars in hx.containers:
    hx.bar_label(bars)
```

```python
sns.set(rc={"figure.figsize":(20,5)})
ix = df.groupby(["Occupation","Gender"],as_index = False)["Amount"].sum().sort_values(by ="Amount" ,ascending = False)
#print(ix)
sns.barplot(data =ix ,x ="Occupation" , y="Amount",hue ="Gender")

```

### Product Category

```python
sns.set(rc ={"figure.figsize":(30,5)})
sns.countplot(data =df ,x ="Product_Category",hue ="Product_Category")
```

```python
sns.set(rc ={"figure.figsize":(25,10)})
kx  =df.groupby(["Product_Category"],as_index =False)["Amount"].sum().sort_values(by ="Amount",ascending =False).head(10)
print(kx)
sns.barplot(data =kx ,x = "Product_Category", y ="Amount",hue ="Product_Category")
```

#### From above graphs we find that most sold products are Food ,Clothing & Apparel and Electronics& Gadgets.

```python
# top 10 most sold products 
lx =df.groupby(["Product_ID"],as_index =False)["Amount"].sum().sort_values(by ="Amount",ascending =False).head(10)
sns.set(rc ={"figure.figsize":(25,10)})
sns.barplot(data =lx ,x ="Product_ID",y ="Amount",hue  ="Product_ID")
```

### Conclusion 

#### Married women age group 25-35 years from UP ,Maharastra and Karnataka working in IT,Healthcare and Aviation are more likely to buy products from Food ,Clothing and Electronics category.
